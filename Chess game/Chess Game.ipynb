```python
def pawn_move(chess_board, a, b, c, d):
    if chess_board[a][b] == 'BP':
        if c == a + 1 and d == b and chess_board[c][d] == '  ':
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            if c == 7:
                promotion_needed = True
            return True
        elif  a == 1 and c == 3 and d == b and chess_board[2][b] == '  ' and chess_board[3][b] == '  ':
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '  
            return True
        elif c == a + 1 and chess_board[c][d] != '  ' and (d == b + 1 or d == b - 1):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            if c == 7:
                promotion_needed = True
            return True
        else:
            return False
    elif chess_board[a][b] == 'WP':
        if c == a - 1 and d == b and chess_board[c][d] == '  ':
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            if c == 0:
                promotion_needed = True
            return True
        elif a == 6 and c == 4 and d == b and chess_board[5][b] == '  ' and chess_board[4][b] == '  ':
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            return True
        elif c == a - 1 and chess_board[c][d] != '  ' and (d == b + 1 or d == b - 1):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            if c == 0:
                promotion_needed = True
            return True
        else:
            return False
    else:
        return 'Other'
    return 'End'
```

```python
def knight_move(chess_board, a, b, c, d):
    if 'N' in chess_board[a][b]:
        if c == a + 2 and (d == b - 1 or d == b + 1):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            return True
        elif c == a + 1 and (d == b - 2 or d == b + 2):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            return True
        elif c == a - 1 and (d == b - 2 or d == b + 2):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            return True
        elif c == a - 2 and (d == b - 1 or d == b + 1):
            chess_board[c][d] = chess_board[a][b]
            chess_board[a][b] = '  '
            return True
        else:
            return False
    else:
        return 'Other'
```

```python
def check_in_between(chess_board, a, b, c, d):
    coordinates = []
    if 'R' in chess_board[a][b]:
        if b == d and c > a:
            for i in range(a + 1, c):
                if chess_board[i][b] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b])
            return [True, coordinates]
        elif b == d and c < a:
            for i in range(c + 1, a):
                if chess_board[i][b] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i,b])
            return [True, coordinates]
        elif a == c and d > b:
            for i in range(b + 1, d):
                if chess_board[a][i] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([a,i])
            return [True, coordinates]
        elif a == c and d < b:
            for i in range(d + 1, b):
                if chess_board[a][i] != '  ':
                    return [False,[[8, 8]]]
                coordinates.append([a,i])
            return [True, coordinates]
        else:
            return [False, [[8, 8]]]
        
    elif 'I' in chess_board[a][b]:
        if b < d and a < c and d - b == c - a:
            for i in range(a + 1, c):
                if chess_board[i][b + i - a] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b+i-a])
            return [True, coordinates]
        elif b < d and a > c:
            for i in range(c + 1, a):
                if chess_board[i][d - i + c] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, d-i+c])
            return [True, coordinates]
        elif b > d and a < c:
            for i in range(a + 1, c):
                if chess_board[i][b - i + a] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b-i+a])
            return [True, coordinates]
        elif b > d and a > c:
            for i in range(c + 1, a):
                if chess_board[i][d + i - c] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, d+i-c])
            return [True, coordinates]
        else:
            return [False, [[8, 8]]]
        
    elif 'Q' in chess_board[a][b]:
        if b == d and c > a:
            for i in range(a + 1, c):
                if chess_board[i][b] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b])
            return [True, coordinates]
        elif b == d and c < a:
            for i in range(c + 1, a):
                if chess_board[i][b] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b])
            return [True, coordinates]
        elif a == c and d > b:
            for i in range(b + 1, d):
                if chess_board[a][i] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([a,i])
            return [True, coordinates]
        elif a == c and d < b:
            for i in range(d + 1, b):
                if chess_board[a][i] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([a,i])
            return [True, coordinates]
        elif b < d and a < c and abs(d - b) == abs(c - a):
            for i in range(a + 1, c):
                if chess_board[i][b + i - a] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b+i-a])
            return [True, coordinates]
        elif b < d and a > c and abs(d - b) == abs(c - a):
            for i in range(c + 1, a):
                if chess_board[i][d - i + c] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, d-i+c])
            return [True, coordinates]
        elif b > d and a < c and abs(d - b) == abs(c - a):
            for i in range(a + 1, c):
                if chess_board[i][b - i + a] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, b-i+a])
            return [True, coordinates]
        elif b > d and a > c and abs(d - b) == abs(c - a):
            for i in range(c + 1, a):
                if chess_board[i][d + i - c] != '  ':
                    return [False, [[8, 8]]]
                coordinates.append([i, d+i-c])
            return [True, coordinates]
        else:
            return [False, [[8, 8]]]
        
    elif 'K' in chess_board[a][b]:
            if d == 2 and 'BK' in chess_board[a][b] and 'BR' in chess_board[0][0]:
                for i in range(1, 4):
                    if chess_board[0][i] != '  ':
                        return [False, [[8, 8]]]
                return [True, [[8, 8]]]
            elif d == 6 and 'BK' in chess_board[a][b] and 'BR' in chess_board[0][7]:
                for i in range(5, 7):
                    if chess_board[0][i] != '  ':
                        return [False, [[8, 8]]]
                return [True, [[8, 8]]]
            elif d == 2 and 'WK' in chess_board[a][b] and 'WR' in chess_board[7][0]:
                for i in range(1, 4):
                    if chess_board[7][i] != '  ':
                        return [False, [[8, 8]]]
                return [True, [[8, 8]]]
            elif d == 6 and 'WK' in chess_board[a][b] and 'WR' in chess_board[7][7]:
                for i in range(5, 7):
                    if chess_board[7][i] != '  ':
                        return [False, [[8, 8]]]
                return [True, [[8, 8]]]
            else:
                return [False, [[8, 8]]]
    else:
        # what to do if not Rook, Bishop, Queen, or King
        return [False, [[8, 8]]]
    
    
    
```

```python
def check(chess_board, king_rook_moved):
    white_king_a = 0
    white_king_b = 0
    
    black_king_a = 0
    black_king_b = 0
    
    for i in range(0, 8):
        for j in range(0, 8):
            if chess_board[i][j] == 'WK':
                white_king_a = i
                white_king_b = j
            if chess_board[i][j] == 'BK':
                black_king_a = i
                black_king_b = j
        
    # White King check captures
    
    # check if pawn can capture king
    
    if ((white_king_a - 1) >= 0 and (white_king_a - 1) <= 7 and (white_king_b - 1) >= 0 and (white_king_b - 1) <= 7) and (chess_board[white_king_a - 1][white_king_b - 1] == 'BP'):
        return [True, white_king_a - 1, white_king_b - 1, 'black']
    if ((white_king_a - 1) >= 0 and (white_king_a - 1) <= 7 and (white_king_b + 1) >= 0 and (white_king_b + 1) <= 7) and (chess_board[white_king_a - 1][white_king_b + 1] == 'BP'):
        return [True, white_king_a - 1, white_king_b + 1, 'black']
                                                                                                                          
    
    
    # check if rook can capture king
    for i in range(0, white_king_a):
        if chess_board[i][white_king_b] == 'BR' and check_in_between(chess_board, i, white_king_b, white_king_a, white_king_b)[0]:
            return [True, i, white_king_b, 'black']
    for i in range(white_king_a + 1, 8):
        if chess_board[i][white_king_b] == 'BR' and check_in_between(chess_board, i, white_king_b, white_king_a, white_king_b)[0]:
            return [True, i, white_king_b, 'black']
    for i in range(0, white_king_b):
        if chess_board[white_king_a][i] == 'BR' and check_in_between(chess_board, white_king_a, i, white_king_a, white_king_b)[0]:
            return [True, white_king_a, i, 'black']
    for i in range(white_king_b + 1, 8):
        if chess_board[white_king_a][i] == 'BR' and check_in_between(chess_board, white_king_a, i, white_king_a, white_king_b)[0]:
            return [True, white_king_a, i, 'black']
        
    # check if knight can capture king
    
    
    if (white_king_a + 2 >= 0 and white_king_a + 2 <= 7 and white_king_b - 1 >= 0 and white_king_b - 1 <= 7) and chess_board[white_king_a + 2][white_king_b - 1] == 'BN':
        return [True, white_king_a + 2, white_king_b - 1, 'black']
    if (white_king_a + 2 >= 0 and white_king_a + 2 <= 7 and white_king_b + 1 >= 0 and white_king_b + 1 <= 7) and chess_board[white_king_a + 2][white_king_b + 1] == 'BN':
        return [True, white_king_a + 2, white_king_b + 1, 'black']
    if (white_king_a + 1 >= 0 and white_king_a + 1 <= 7 and white_king_b - 2 >= 0 and white_king_b - 2 <= 7) and chess_board[white_king_a + 1][white_king_b - 2] == 'BN':
        return [True, white_king_a + 1, white_king_b - 2, 'black']
    if (white_king_a + 1 >= 0 and white_king_a + 1 <= 7 and white_king_b + 2 >= 0 and white_king_b + 2 <= 7) and chess_board[white_king_a + 1][white_king_b + 2] == 'BN':
        return [True, white_king_a + 1, white_king_b + 2, 'black']
    if (white_king_a - 1 >= 0 and white_king_a - 1 <= 7 and white_king_b - 2 >= 0 and white_king_b - 2 <= 7) and chess_board[white_king_a - 1][white_king_b - 2] == 'BN':
        return [True, white_king_a - 1, white_king_b - 2, 'black']
    if (white_king_a - 1 >= 0 and white_king_a - 1 <= 7 and white_king_b + 2 >= 0 and white_king_b + 2 <= 7) and chess_board[white_king_a - 1][white_king_b + 2] == 'BN':
        return [True, white_king_a - 1, white_king_b + 2, 'black']
    if (white_king_a - 2 >= 0 and white_king_a - 2 <= 7 and white_king_b - 1 >= 0 and white_king_b - 1 <= 7) and chess_board[white_king_a - 2][white_king_b - 1] == 'BN':
        return [True, white_king_a - 2, white_king_b - 1, 'black']
    if (white_king_a - 2 >= 0 and white_king_a - 2 <= 7 and white_king_b + 1 >= 0 and white_king_b + 1 <= 7) and chess_board[white_king_a - 2][white_king_b + 1] == 'BN':
        return [True, white_king_a - 2, white_king_b + 1, 'black']

    # check if bishop can capture king
    
    for i in range(0, 8):
        for j in range(0, 8):
            if chess_board[i][j] == 'BI' and abs(j - white_king_b) == abs(i - white_king_a) and check_in_between(chess_board, i, j, white_king_a, white_king_b)[0]:
                return [True, i, j, 'black']
            
    # check if Queen can capture King
    for i in range(0, 8):
        for j in range(0, 8):
            if chess_board[i][j] == 'BQ' and abs(j - white_king_b) == abs(i - white_king_a) and check_in_between(chess_board, i, j, white_king_a, white_king_b)[0]:
                return [True, i, j, 'black']
        
    for i in range(0, white_king_a):
        if chess_board[i][white_king_b] == 'BQ' and check_in_between(chess_board, i, white_king_b, white_king_a, white_king_b)[0]:
            return [True, i, white_king_b, 'black']
    for i in range(white_king_a + 1, 8):
        if chess_board[i][white_king_b] == 'BQ' and check_in_between(chess_board, i, white_king_b, white_king_a, white_king_b)[0]:
            return [True, i, white_king_b, 'black']
    for i in range(0, white_king_b):
        if chess_board[white_king_a][i] == 'BQ' and check_in_between(chess_board, white_king_a, i, white_king_a, white_king_b)[0]:
            return [True, white_king_a, i, 'black']
    for i in range(white_king_b + 1, 8):
        if chess_board[white_king_a][i] == 'BQ' and check_in_between(chess_board, white_king_a, i, white_king_a, white_king_b)[0]:
            return [True, white_king_a, i, 'black']
        
    
    # Black King check captures
    
    # check if pawn can capture king
    
    if ((black_king_a + 1) >= 0 and (black_king_a + 1) <= 7 and (black_king_b - 1) >= 0 and (black_king_b - 1) <= 7) and (chess_board[black_king_a + 1][black_king_b - 1] == 'WP'):
        return [True, black_king_a +1, black_king_b -1 , 'white']
    if ((black_king_a + 1) >= 0 and (black_king_a + 1) <= 7 and (black_king_b + 1) >= 0 and (black_king_b + 1) <= 7) and (chess_board[black_king_a + 1][black_king_b + 1] == 'WP'):
        return [True, black_king_a + 1, black_king_b + 1, 'white']
    
    # check if rook can capture king
    for i in range(0, black_king_a):
        if chess_board[i][black_king_b] == 'WR' and check_in_between(chess_board, i, black_king_b, black_king_a, black_king_b)[0]:
            return [True, i, black_king_b, 'white']
    for i in range(black_king_a + 1, 8):
        if chess_board[i][black_king_b] == 'WR' and check_in_between(chess_board, i, black_king_b, black_king_a, black_king_b)[0]:
            return [True, i, black_king_b, 'white']
    for i in range(0, black_king_b):
        if chess_board[black_king_a][i] == 'WR' and check_in_between(chess_board, black_king_a, i, black_king_a, black_king_b)[0]:
            return [True, black_king_a, i, 'white']
    for i in range(black_king_b + 1, 8):
        if chess_board[black_king_a][i] == 'WR' and check_in_between(chess_board, black_king_a, i, black_king_a, black_king_b)[0]:
            return [True, black_king_a, i, 'white']
        
    # check if knight can capture king
    
    
    if (black_king_a + 2 >= 0 and black_king_a + 2 <= 7 and black_king_b - 1 >= 0 and black_king_b - 1 <= 7) and chess_board[black_king_a + 2][black_king_b - 1] == 'WN':
        return [True, black_king_a + 2, black_king_b - 1, 'white']
    if (black_king_a + 2 >= 0 and black_king_a + 2 <= 7 and black_king_b + 1 >= 0 and black_king_b + 1 <= 7) and chess_board[black_king_a + 2][black_king_b + 1] == 'WN':
        return [True, black_king_a + 2, black_king_b + 1, 'white']
    if (black_king_a + 1 >= 0 and black_king_a + 1 <= 7 and black_king_b - 2 >= 0 and black_king_b - 2 <= 7) and chess_board[black_king_a + 1][black_king_b - 2] == 'WN':
        return [True, black_king_a + 1, black_king_b - 2, 'white']
    if (black_king_a + 1 >= 0 and black_king_a + 1 <= 7 and black_king_b + 2 >= 0 and black_king_b + 2 <= 7) and chess_board[black_king_a + 1][black_king_b + 2] == 'WN':
        return [True, black_king_a + 1, black_king_b + 2, 'white']
    if (black_king_a - 1 >= 0 and black_king_a - 1 <= 7 and black_king_b - 2 >= 0 and black_king_b - 2 <= 7) and chess_board[black_king_a - 1][black_king_b - 2] == 'WN':
        return [True, black_king_a - 1, black_king_b - 2, 'white']
    if (black_king_a - 1 >= 0 and black_king_a - 1 <= 7 and black_king_b + 2 >= 0 and black_king_b + 2 <= 7) and chess_board[black_king_a - 1][black_king_b + 2] == 'WN':
        return [True, black_king_a -1, black_king_b + 2, 'white']
    if (black_king_a - 2 >= 0 and black_king_a - 2 <= 7 and black_king_b - 1 >= 0 and black_king_b - 1 <= 7) and chess_board[black_king_a - 2][black_king_b - 1] == 'WN':
        return [True, black_king_a - 2, black_king_b - 1, 'white']
    if (black_king_a - 2 >= 0 and black_king_a - 2 <= 7 and black_king_b + 1 >= 0 and black_king_b + 1 <= 7) and chess_board[black_king_a - 2][black_king_b + 1] == 'WN':
        return [True, black_king_a - 2, black_king_b + 1, 'white']

    # check if bishop can capture king
    
    for i in range(0, 8):
        for j in range(0, 8):
            if chess_board[i][j] == 'WI' and abs(j - black_king_b) == abs(i - black_king_a) and check_in_between(chess_board, i, j, black_king_a, black_king_b)[0]:
                return [True, i, j, 'white']
            
    # check if Queen can capture King
    for i in range(0, 8):
        for j in range(0, 8):
            if chess_board[i][j] == 'WQ' and abs(j - black_king_b) == abs(i - black_king_a) and check_in_between(chess_board, i, j, black_king_a, black_king_b)[0]:
                return [True, i, j, 'white']
        
    for i in range(0, black_king_a):
        if chess_board[i][black_king_b] == 'WQ' and check_in_between(chess_board, i, black_king_b, black_king_a, black_king_b)[0]:
            return [True, i, black_king_b, 'white']
    for i in range(black_king_a + 1, 8):
        if chess_board[i][black_king_b] == 'WQ' and check_in_between(chess_board, i, black_king_b, black_king_a, black_king_b)[0]:
            return [True, i, black_king_b, 'white']
    for i in range(0, black_king_b):
        if chess_board[black_king_a][i] == 'WQ' and check_in_between(chess_board, black_king_a, i, black_king_a, black_king_b)[0]:
            return [True, black_king_a, i, 'white']
    for i in range(black_king_b + 1, 8):
        if chess_board[black_king_a][i] == 'WQ' and check_in_between(chess_board, black_king_a, i, black_king_a, black_king_b)[0]:
            return [True, black_king_a, i, 'white']
        
    return [False, False, False, False]
```
