```python
# COGS
COGS2023 = IncomeStatement2024.iloc[34].at['2023']
COGS2022 = IncomeStatement2023.iloc[35].at['2022']
COGS2021 = IncomeStatement2023.iloc[34].at['2021']
```

```python
# Revennue
Rev2023 = IncomeStatement2024.iloc[3].at['2023']
Rev2022 = IncomeSheet2023.iloc[3].at['2022']
Rev2021 = IncomeSheet2022.iloc[3].at['2021']
```

```python
# Net Income
NetIncome2023 = IncomeStatement2024.iloc[21].at['2023'] * IncomeStatement2024.iloc[22].at['2023'] / 1000000
NetIncome2022 = IncomeStatement2023.iloc[22].at['2022'] * IncomeStatement2023.iloc[23].at['2022'] / 1000000
NetIncome2021 = IncomeStatement2022.iloc[21].at['2021'] * IncomeStatement2022.iloc[22].at['2021'] / 1000000
```

```python
# Total Assets
TA2023 = BalanceSheet2024.iloc[8].at['2023']
TA2022 = BalanceSheet2023.iloc[8].at['2022']
TA2021 = BalanceSheet2022.iloc[8].at['2021']
```

```python
# Shareholder's Equity
SE2023 = BalanceSheet2024.iloc[22].at['2023']
SE2022 = BalanceSheet2023.iloc[20].at['2022']
SE2021 = BalanceSheet2022.iloc[22].at['2021']
```

```python
financial_data = pd.DataFrame([[2021, COGS2021, Rev2021, NetIncome2021, TA2021, SE2021], 
                               [2022, COGS2022, Rev2022, NetIncome2022, TA2022, SE2022], 
                               [2023, COGS2023, Rev2023, NetIncome2023, TA2023, SE2023]], columns = ['Year', 'COGS', 'Revenue', 'Net Income', 'Total Assets', 'Shareholder\'s Equity'])
```

```python
financial_data
```

```python
# Profitability Ratios

# Gross Profit Margin = (Revenue - COGs) / Revenue

financial_data['Gross Profit Margin'] = (financial_data['Revenue'] - financial_data['COGS']) / financial_data['Revenue']

# Net Profit Margin = Net Income / Revenue
financial_data['Net Profit Margin'] = financial_data['Net Income'] / financial_data['Revenue']

# Return on Assets (ROA) = Net Income / Total Assets
financial_data['ROA'] = financial_data['Net Income'] / financial_data['Total Assets']

# Return on Equity (ROE) = Net Income / Shareholder's Equity
financial_data['ROE'] = financial_data['Net Income'] / financial_data['Shareholder\'s Equity']
```

```python
financial_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
plt.figure(figsize=(10, 5))
plt.plot(financial_data['Year'], financial_data['Revenue'], marker='o', label='Revenue')
plt.plot(financial_data['Year'], financial_data['Net Income'], marker='o', label='Net Income')
plt.title('Revenue and Net Income Trend')
plt.xlabel('Year')
plt.xticks(financial_data['Year'])
plt.ylabel('Amount (in millions)')
plt.legend()
plt.show()
```
